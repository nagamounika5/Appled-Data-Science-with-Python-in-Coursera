```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
import seaborn as sns
%matplotlib notebook

plt.style.use('seaborn-colorblind')

#df = pd.read_csv('data/C2A2_data/BinnedCsvs_d25/9bc594d0d6bf5fec16beb2afb02a3b859b7d804548c77d614b2a6b9b.csv')
df = pd.read_csv('data/C2A2_data/BinnedCsvs_d25/000899796af1c692527b8a5a94629103aba7a1e1657388df56c761da.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/datasets/global-temp/master/data/monthly.csv')
df = df.sort(['ID', 'Date'])

df2 = df2.set_index('Date')
df2 = df2[df2['Source'] == 'GCAG']
df2 = df2.reset_index()

df2['Year'] = df2['Date'].apply(lambda x: x[:4]) 
df2['Month'] = df2['Date'].apply(lambda x: x[-2:])

df2 = df2[df2['Year'] >= '2000']

avg = []
for i in range(1,10):
    climate = df2[df2['Month'] == '0'+str(i)].median()
    avg.append(climate)

for i in range(10,13):
    climate = df2[df2['Month'] == str(i)].median()
    avg.append(climate)

df['Year'] = df['Date'].apply(lambda x: x[:4])
df['Month-Day'] = df['Date'].apply(lambda x: x[5:])
df = df[df['Month-Day'] != '02-29']

df_min = df[(df['Element'] == 'TMIN')]
df_max = df[(df['Element'] == 'TMAX')]

# df_min = df[(df['Element'] == 'TMIN')]
# df_max = df[(df['Element'] == 'TMAX')]

min1 = df_min.groupby('Month-Day')['Data_Value'].agg({'temp_min_mean_NewYork': np.median}).reset_index()
max1 = df_max.groupby('Month-Day')['Data_Value'].agg({'temp_max_mean_NewYork': np.median}).reset_index()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(min1['temp_min_mean_NewYork']/10, 'y', alpha = 0.25, label = 'Low Temperature in New York')
ax1.plot(max1['temp_max_mean_NewYork']/10, 'r', alpha = 0.75, label = 'High Temperature in New York')


a = [0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330]
b = [i+15 for i in a]
ax2.plot(b, avg, 'g', alpha = 0.75, label = 'Global Surface')
ax1.set_ylabel('New York')
ax2.set_ylabel('Global Ocean Surface')
plt.title('Camparing the Climate of New York \n against the Global Surface of Ocean ')

ax1.legend()
ax2.legend()

month = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
plt.xticks(b, month)

for each in ax1.spines.values():
    each.set_visible(False)
for each in ax2.spines.values():
    each.set_visible(False)

plt.show()
```
